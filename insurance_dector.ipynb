```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/insurance.csv")
# data.head()
data.loc[data["sex"]=="female",'sex']=0
data.loc[data["sex"]=="male",'sex']=1
data.loc[data["smoker"]=="yes",'smoker']=0
data.loc[data["smoker"]=="no",'smoker']=1
data.loc[data["region"]=="southwest",'region']=0
data.loc[data["region"]=="southeast",'region']=1
data.loc[data["region"]=="northwest",'region']=2
data.loc[data["region"]=="northeast",'region']=3
# data.replace({"region": {'southeast': 0, 'southwest': 1, 'northeast': 2, 'northwest': 3}})

data
```

```python
x=data.drop(columns="charges",axis=1)
y=data["charges"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=1,train_size=0.3)
```

```python
x_train
```

```python
model=LinearRegression()
model_=model.fit(x_train,y_train)
predict=model.predict(x_train)
accuracyscore=metrics.r2_score(y_train,predict)
accuracyscore
```

```python
predict=model.predict(x_test)
accuracyscore=metrics.r2_score(y_test,predict)
accuracyscore
```
